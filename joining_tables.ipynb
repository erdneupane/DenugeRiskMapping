```python
import pandas as pd

# -----------------------------
# 1. Load your two CSV files
# -----------------------------
env_df = pd.read_csv(r"C:\Users\dhrub\Desktop\Wards_Monthly_Aggregated_2019_2023.csv")          # left file (NDVI, EVI, LST, precipitation)
dengue_df = pd.read_csv(r"C:\Users\dhrub\Desktop\Dengue_Ward_Monthly_Aggregated.csv") # right file (dengue_cases)
# -----------------------------
# 2. Convert merge columns to same datatype
# -----------------------------
# Strip spaces from names
env_df['name'] = env_df['name'].astype(str).str.strip()
dengue_df['name'] = dengue_df['name'].astype(str).str.strip()

# Convert year & month to integers in both dataframes
env_df['year'] = env_df['year'].astype(int)
env_df['month'] = env_df['month'].astype(int)

dengue_df['year'] = dengue_df['year'].astype(int)
dengue_df['month'] = dengue_df['month'].astype(int)

# -----------------------------
# 3. Merge both datasets
# -----------------------------
merged_df = pd.merge(
    env_df,
    dengue_df[['name', 'year', 'month', 'dengue_cases']],
    how='left',                # keep all environmental data rows
    on=['name', 'year', 'month']
)

# -----------------------------
# 4. Save result
# -----------------------------
merged_df.to_csv("Merged_Ward_Monthly_Data.csv", index=False)

print("Merging completed! Saved as Merged_Ward_Monthly_Data.csv")
```
