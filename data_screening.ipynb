```python
# ================================
# 1. IMPORT LIBRARIES
# ================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ================================
# 2. LOAD YOUR CSV FILE
# ================================
file_path = r"C:\Users\dhrub\Desktop\Data\kaski_env_dengue_merged.csv"

df = pd.read_csv(file_path)

# ================================
# 3. SELECT ENVIRONMENTAL VARIABLES
# ================================
env_cols = ['Precip', 'NDVI', 'EVI', 'LST_Night', 'LST_Day']

env_df = df[env_cols].dropna()

# Rename (paper style)
env_df = env_df.rename(columns={
    'Precip': 'Precipitation',
    'LST_Night': 'nLST',
    'LST_Day': 'dLST'
})

# ================================
# 4. CORRELATION MATRIX
# ================================
corr_matrix = env_df.corr(method='pearson').round(3)

print("\n===== Pearson Correlation Table =====\n")
print(corr_matrix)

# Save table
corr_matrix.to_csv("correlation_table.csv")

# ================================
# 5. HEATMAP + SAVE
# ================================
plt.figure(figsize=(7,6))

sns.heatmap(corr_matrix, annot=True)

plt.title("Correlation Heatmap (Environmental Variables)")
plt.tight_layout()

# 👉 SAVE IMAGE (IMPORTANT LINE)
plt.savefig("correlation_heatmap.png", dpi=300)

plt.show()
```
