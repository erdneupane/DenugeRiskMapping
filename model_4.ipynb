```python
# ============================================================
# Dengue Prediction Methodology — Kaski District, Nepal
# Following Acharya et al. (2018) — Chitwan District Framework
# Models: Poisson GAM (primary) + RF + SVM (comparative)
# ============================================================
# Reference:
# Acharya, B.K., Cao, C., Xu, M., Khanal, L., Naeem, S.,
# & Pandit, S. (2018). Temporal Variations and Associated
# Remotely Sensed Environmental Variables of Dengue Fever
# in Chitwan District, Nepal. ISPRS Int. J. Geo-Inf., 7, 275.
# ============================================================
# Requirements:
#   pip install pygam scikit-learn statsmodels scipy
#               pandas numpy matplotlib
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import statsmodels.api as sm
import warnings
from itertools import combinations
from scipy import stats
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import (mean_squared_error,
                             mean_absolute_error,
                             r2_score)
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

warnings.filterwarnings("ignore")

# ── pygam import with version diagnostic ─────────────────────
try:
    from pygam import PoissonGAM, s, l
    import pygam as _pygam_mod
    PYGAM_OK      = True
    PYGAM_VERSION = getattr(_pygam_mod, "__version__", "unknown")
    print(f"pygam found — version: {PYGAM_VERSION}")
except ImportError:
    PYGAM_OK = False
    print("WARNING: pygam not installed.  Run:  pip install pygam")


# ============================================================
# GLOBAL SETTINGS — edit to match your file / column names
# ============================================================

DATA_PATH     = r"C:\Users\dhrub\Desktop\Data\kaski_env_dengue_merged.csv"

CASE_COL      = "dengue_cases"
YEAR_COL      = "year"
MONTH_COL     = "month"

ENV_COLS      = ["NDVI", "EVI", "LST_Day", "LST_Night", "Precip"]

MAX_LAG       = 6
MULTICOLL_THR = 0.7
N_CV_SPLITS   = 10
N_SPLINES     = 4
RANDOM_STATE  = 42


# ============================================================
# HELPER — robust pygam statistics extractor
# Works across pygam 0.8.x and 0.9.x which use different keys
# ============================================================

def _safe_gam_stats(gam):
    """
    Extract (AIC, deviance_explained_%, pseudo_R2) from a fitted
    PoissonGAM regardless of pygam version.

    pygam 0.8.x uses:
        statistics_['AICc'], ['deviance'], ['null_deviance'],
        ['pseudo_r2']['McFadden']

    pygam 0.9.x / some builds use:
        statistics_['AIC'], ['loglikelihood'],
        ['null_deviance'] may be absent — we reconstruct it.
    """
    st = gam.statistics_

    # ── AIC ──────────────────────────────────────────────────
    aic = (st.get("AIC")
           or st.get("AICc")
           or st.get("aic")
           or st.get("aicc")
           or np.nan)

    # ── deviance & null deviance ──────────────────────────────
    dev = (st.get("deviance")
           or st.get("Deviance")
           or np.nan)

    null_dev = (st.get("null_deviance")
                or st.get("NullDeviance")
                or st.get("null deviance")
                or None)

    # If null_deviance missing, reconstruct from log-likelihood
    if null_dev is None:
        ll   = st.get("loglikelihood") or st.get("LogLikelihood")
        ll0  = st.get("null_loglikelihood") or st.get("NullLogLikelihood")
        if ll is not None and ll0 is not None:
            null_dev = 2 * (ll0 - ll) + dev   # approximate
        else:
            null_dev = dev + 1e-9              # fallback

    dev_exp = (1 - dev / null_dev) * 100 if null_dev else 0.0

    # ── pseudo R² ────────────────────────────────────────────
    pr2_block = st.get("pseudo_r2") or {}
    if isinstance(pr2_block, dict):
        pr2 = (pr2_block.get("McFadden")
               or pr2_block.get("mcfadden")
               or pr2_block.get("Mcfadden")
               or list(pr2_block.values())[0]
               if pr2_block else np.nan)
    else:
        pr2 = float(pr2_block) if pr2_block else np.nan

    # If pseudo_r2 still missing, compute from deviances
    if pr2 is None or np.isnan(pr2):
        pr2 = 1 - dev / null_dev if null_dev else np.nan

    return float(aic), float(dev_exp), float(pr2)


def _print_stats_keys(gam):
    """Print all available statistics_ keys — for debugging."""
    st = gam.statistics_
    print("\n  [DEBUG] statistics_ keys available in your pygam:")
    for k, v in st.items():
        print(f"    {k!r}: {v}")
    print()


# ============================================================
# HELPER FUNCTIONS
# ============================================================

def build_gam_terms(n_predictors):
    """
    Column 0 = time spline.
    Columns 1 .. n_predictors = linear environmental predictors.
    """
    terms = s(0, n_splines=N_SPLINES, spline_order=3)
    for k in range(1, n_predictors + 1):
        terms = terms + l(k)
    return terms


def fit_gam(t_col, X_pred, y):
    """
    Fit PoissonGAM.
    Returns (gam, X_full_matrix).
    """
    if X_pred.ndim == 1:
        X_pred = X_pred.reshape(-1, 1)
    X_full = np.column_stack([t_col.reshape(-1, 1), X_pred])
    gam    = PoissonGAM(build_gam_terms(X_pred.shape[1]))
    gam.fit(X_full, y)
    return gam, X_full


def regression_metrics(y_true, y_pred):
    yt = np.array(y_true,  dtype=float)
    yp = np.clip(np.array(y_pred, dtype=float), 0, None)
    r, _ = stats.pearsonr(yt, yp)
    return {
        "RMSE"     : float(np.sqrt(mean_squared_error(yt, yp))),
        "MAE"      : float(mean_absolute_error(yt, yp)),
        "R2"       : float(r2_score(yt, yp)),
        "Pearson_r": float(r)
    }


def save_fig(name):
    plt.savefig(name, dpi=300, bbox_inches="tight")
    print(f"  Saved: {name}")


# ============================================================
# SECTION 1 — LOAD AND PREPARE DATA
# ============================================================

print("\n" + "=" * 65)
print("DENGUE PREDICTION — KASKI DISTRICT, NEPAL")
print("Following Acharya et al. (2018)")
print("=" * 65)

df = pd.read_csv(DATA_PATH)
df = df.sort_values([YEAR_COL, MONTH_COL]).reset_index(drop=True)
df["date"]       = pd.to_datetime(
    df[[YEAR_COL, MONTH_COL]].assign(day=1)
)
df["time_index"] = np.arange(len(df), dtype=float)

env_vars = [v for v in ENV_COLS if v in df.columns]
missing  = [v for v in ENV_COLS if v not in df.columns]
if missing:
    print(f"\n  Columns not found (skipped): {missing}")

print(f"\n  Rows       : {len(df)}")
print(f"  Period     : {df['date'].min():%b %Y} – "
      f"{df['date'].max():%b %Y}")
print(f"  Outcome    : {CASE_COL}")
print(f"  Predictors : {env_vars}")


# ============================================================
# SECTION 2 — DESCRIPTIVE STATISTICS
# ============================================================

print("\n" + "=" * 65)
print("SECTION 2: DESCRIPTIVE STATISTICS")
print("=" * 65)

stat_cols = [CASE_COL] + env_vars
desc = df[stat_cols].describe(
    percentiles=[0.05, 0.25, 0.50, 0.75, 0.95, 0.99]
).T
desc.columns = ["n", "Mean", "SD", "Min",
                "5%", "25%", "50%", "75%", "95%", "99%", "Max"]
desc = desc.drop(columns=["SD"])
print("\n", desc.round(3).to_string())
desc.to_csv("descriptive_statistics.csv")
print("\n  Saved: descriptive_statistics.csv")


# ============================================================
# SECTION 3 — TIME SERIES VISUALIZATION
# ============================================================

print("\n" + "=" * 65)
print("SECTION 3: TIME SERIES PLOTS")
print("=" * 65)

pal  = ["red", "green", "darkgreen", "darkorange", "purple", "steelblue"]
ylab = {
    CASE_COL   : "Cases",
    "NDVI"     : "NDVI",
    "EVI"      : "EVI",
    "LST_Day"  : "dLST (°C)",
    "LST_Night": "nLST (°C)",
    "Precip"   : "Precip (mm)"
}

n   = len(stat_cols)
fig, axes = plt.subplots(n, 1, figsize=(12, 2.5 * n),
                          dpi=150, sharex=True)
if n == 1:
    axes = [axes]

for ax, col, c in zip(axes, stat_cols, pal):
    ax.plot(df["date"], df[col], color=c, linewidth=1.2)
    ax.set_ylabel(ylab.get(col, col), fontsize=9)
    ax.grid(True, alpha=0.3)
    ax.xaxis.set_major_locator(mdates.YearLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))

axes[-1].set_xlabel("Year")
fig.suptitle("Monthly Time Series — Kaski District",
             fontsize=11, fontweight="bold")
plt.tight_layout()
save_fig("timeseries_plot.png")
plt.show()


# ============================================================
# SECTION 4 — INTER-PREDICTOR CORRELATION MATRIX
# ============================================================

print("\n" + "=" * 65)
print("SECTION 4: INTER-PREDICTOR CORRELATION MATRIX")
print("=" * 65)

corr_mat = df[env_vars].corr(method="pearson")
print("\nPearson r matrix:")
print(corr_mat.round(3).to_string())

excluded_pairs = set()
print(f"\nPairs with |r| > {MULTICOLL_THR} — excluded from same model:")
flagged = False
for i in range(len(env_vars)):
    for j in range(i + 1, len(env_vars)):
        r_val = corr_mat.iloc[i, j]
        if abs(r_val) > MULTICOLL_THR:
            v1, v2 = env_vars[i], env_vars[j]
            excluded_pairs.update([(v1, v2), (v2, v1)])
            print(f"  {v1} <-> {v2} : r = {r_val:.3f}  >> EXCLUDED")
            flagged = True
if not flagged:
    print("  None found above threshold.")

corr_mat.to_csv("predictor_correlation_matrix.csv")

fig, ax = plt.subplots(figsize=(7, 5), dpi=150)
im = ax.imshow(corr_mat.values, cmap="RdYlBu_r", vmin=-1, vmax=1)
plt.colorbar(im, ax=ax)
ax.set_xticks(range(len(env_vars)))
ax.set_yticks(range(len(env_vars)))
ax.set_xticklabels(env_vars, rotation=45, ha="right")
ax.set_yticklabels(env_vars)
for i in range(len(env_vars)):
    for j in range(len(env_vars)):
        v = corr_mat.values[i, j]
        ax.text(j, i, f"{v:.2f}", ha="center", va="center",
                fontsize=9,
                color="white" if abs(v) > 0.6 else "black")
ax.set_title(
    "Pearson Correlation Matrix — Predictors\nKaski District"
)
plt.tight_layout()
save_fig("correlation_heatmap_predictors.png")
plt.show()


# ============================================================
# SECTION 5 — CROSS-CORRELATION ANALYSIS (Lags 0–MAX_LAG)
# ============================================================

print("\n" + "=" * 65)
print(f"SECTION 5: CROSS-CORRELATION (Lags 0–{MAX_LAG})")
print("=" * 65)

lag_results = {}

print(f"\n{'Variable':<12} {'Lag':>4}  {'r':>7}  {'p':>9}  {'Sig':>5}")
print("-" * 45)

for var in env_vars:
    rs, ps = [], []
    for lag in range(MAX_LAG + 1):
        y_vals = df[CASE_COL].iloc[lag:].values
        x_vals = df[var].shift(lag).iloc[lag:].values
        valid  = ~(np.isnan(y_vals) | np.isnan(x_vals))
        r, p   = stats.pearsonr(y_vals[valid], x_vals[valid])
        rs.append(r); ps.append(p)
        print(f"  {var:<10} {lag:>4}  {r:>7.3f}  {p:>9.4f}"
              f"  {'Yes*' if p < 0.05 else 'No':>5}")
    lag_results[var] = {"r": rs, "p": ps}
    print()

best_lags = {}
print(f"\n{'Variable':<12} {'BestLag':>8} {'r':>7} {'Sig':>5}")
print("-" * 35)
for var in env_vars:
    rs, ps = lag_results[var]["r"], lag_results[var]["p"]
    best   = int(np.argmax(np.abs(rs)))
    best_lags[var] = best
    print(f"  {var:<10} {best:>8} {rs[best]:>7.3f}"
          f" {'Yes*' if ps[best] < 0.05 else 'No':>5}")

cc_rows = [
    {"Variable": var, "Lag": lag,
     "r"       : round(lag_results[var]["r"][lag], 3),
     "p_value" : round(lag_results[var]["p"][lag], 4),
     "Sig"     : "Yes" if lag_results[var]["p"][lag] < 0.05 else "No"}
    for var in env_vars
    for lag in range(MAX_LAG + 1)
]
pd.DataFrame(cc_rows).to_csv("crosscorrelation_results.csv",
                              index=False)

fig, axes = plt.subplots(1, len(env_vars),
                          figsize=(3.5 * len(env_vars), 4), dpi=150)
if len(env_vars) == 1:
    axes = [axes]

for ax, var in zip(axes, env_vars):
    rs, ps = lag_results[var]["r"], lag_results[var]["p"]
    lgs    = list(range(MAX_LAG + 1))
    cols   = ["tomato" if p < 0.05 else "lightgray" for p in ps]
    ax.bar(lgs, rs, color=cols, edgecolor="black", linewidth=0.5)
    ax.axhline(0,     color="black",     linewidth=0.8)
    ax.axhline( 0.3,  color="steelblue", linestyle="--",
                linewidth=0.7, alpha=0.6)
    ax.axhline(-0.3,  color="steelblue", linestyle="--",
                linewidth=0.7, alpha=0.6)
    best = best_lags[var]
    ax.bar(best, rs[best],
           color="darkred" if ps[best] < 0.05 else "gray",
           edgecolor="black", linewidth=1.2,
           label=f"Best lag = {best}")
    ax.set_xlabel("Lag (months)")
    ax.set_ylabel("Pearson r" if var == env_vars[0] else "")
    ax.set_title(var, fontweight="bold")
    ax.set_xticks(lgs)
    ax.set_ylim(-0.8, 0.8)
    ax.grid(True, alpha=0.3, axis="y")
    ax.legend(fontsize=7)

fig.suptitle(
    "Cross-Correlation: Dengue vs Environmental Variables\n"
    "Red = p<0.05  |  Dark bar = best lag",
    fontsize=10, fontweight="bold"
)
plt.tight_layout()
save_fig("crosscorrelation_plot.png")
plt.show()


# ============================================================
# SECTION 6 — CREATE LAGGED PREDICTOR VARIABLES
# ============================================================

print("\n" + "=" * 65)
print("SECTION 6: LAGGED PREDICTOR VARIABLES")
print("=" * 65)

for var in env_vars:
    lag = best_lags[var]
    df[f"{var}_lag{lag}"] = df[var].shift(lag)
    print(f"  {var}_lag{lag}  (lag = {lag} months)")

df["month_sin"] = np.sin(2 * np.pi * df[MONTH_COL] / 12)
df["month_cos"] = np.cos(2 * np.pi * df[MONTH_COL] / 12)
print("  month_sin, month_cos  (cyclical seasonality)")

df_model = df.dropna().reset_index(drop=True)
print(f"\n  Observations retained: {len(df_model)}")

lag_cols = [f"{v}_lag{best_lags[v]}" for v in env_vars
            if f"{v}_lag{best_lags[v]}" in df_model.columns]


# ============================================================
# SECTION 7 — OVERDISPERSION CHECK
# ============================================================

print("\n" + "=" * 65)
print("SECTION 7: OVERDISPERSION CHECK")
print("=" * 65)

X_ck = sm.add_constant(df_model[lag_cols])
y_ck = df_model[CASE_COL]
p_ck = sm.GLM(y_ck, X_ck,
              family=sm.families.Poisson()).fit(disp=False)
disp = p_ck.deviance / p_ck.df_resid

print(f"\n  Deviance         : {p_ck.deviance:.2f}")
print(f"  df residual      : {p_ck.df_resid:.0f}")
print(f"  Dispersion ratio : {disp:.2f}")
if disp > 1.5:
    print("  >> Overdispersion detected — consider Negative Binomial")
else:
    print("  >> No severe overdispersion — Poisson GAM appropriate")


# ============================================================
# SECTION 8 — UNIVARIATE POISSON GAM MODELS
# ============================================================

print("\n" + "=" * 65)
print("SECTION 8: UNIVARIATE POISSON GAM MODELS")
print("=" * 65)

if not PYGAM_OK:
    print("  SKIPPED — pygam not installed.")
else:
    y_gam    = df_model[CASE_COL].values.astype(float)
    time_idx = df_model["time_index"].values.astype(float)

    print(f"\n  n={len(y_gam)}  min={y_gam.min():.0f}"
          f"  mean={y_gam.mean():.1f}  max={y_gam.max():.0f}"
          f"  NaN={np.isnan(y_gam).sum()}\n")

    # ── Print stats keys once for debugging ──────────────────
    _debug_printed = False

    uni_results = []
    print(f"{'#':<4} {'Variable':<24} {'AIC':>10}"
          f" {'Dev%':>8} {'R2':>8}")
    print("-" * 58)

    for i, lc in enumerate(lag_cols):
        X_u = df_model[[lc]].values.astype(float)
        try:
            gam_u, _ = fit_gam(time_idx, X_u, y_gam)

            # Print keys first time a model fits successfully
            if not _debug_printed:
                _print_stats_keys(gam_u)
                _debug_printed = True

            aic, de, pr2 = _safe_gam_stats(gam_u)
            uni_results.append({
                "Model": f"U{i+1}", "Variable": lc,
                "AIC"      : round(aic,  3),
                "Dev_Expl" : round(de,   1),
                "Pseudo_R2": round(pr2,  3)
            })
            print(f"  U{i+1:<2} {lc:<24} {aic:>10.2f}"
                  f" {de:>7.1f}% {pr2:>8.3f}")
        except Exception as e:
            print(f"  U{i+1:<2} {lc:<24}  FAILED: {e}")

    pd.DataFrame(uni_results).to_csv(
        "univariate_model_results.csv", index=False
    )
    print("\n  Saved: univariate_model_results.csv")


# ============================================================
# SECTION 9 — MULTIVARIATE POISSON GAM MODEL SELECTION
# ============================================================

print("\n" + "=" * 65)
print("SECTION 9: MULTIVARIATE POISSON GAM — MODEL SELECTION")
print("=" * 65)

if not PYGAM_OK:
    print("  SKIPPED — pygam not installed.")
    model_results = []
    best_cols     = lag_cols
    best_res      = None
else:
    avail  = {v: f"{v}_lag{best_lags[v]}"
              for v in env_vars
              if f"{v}_lag{best_lags[v]}" in df_model.columns}
    av_k   = list(avail.keys())

    cand_sets = []
    for k in av_k:
        cand_sets.append([avail[k]])
    for v1, v2 in combinations(av_k, 2):
        if (v1, v2) not in excluded_pairs:
            cand_sets.append([avail[v1], avail[v2]])
    for v1, v2, v3 in combinations(av_k, 3):
        if any(p in excluded_pairs
               for p in [(v1,v2),(v1,v3),(v2,v3)]):
            continue
        cand_sets.append([avail[v1], avail[v2], avail[v3]])

    seen_k, unique_sets = [], []
    for cs in cand_sets:
        key = tuple(sorted(cs))
        if key not in seen_k:
            seen_k.append(key)
            unique_sets.append(cs)

    print(f"\n  {len(unique_sets)} candidate models ...\n")
    print(f"{'#':<5} {'Variables':<58} {'AIC':>10}"
          f" {'Dev%':>7} {'R2':>7}")
    print("-" * 90)

    model_results = []

    for idx, var_set in enumerate(unique_sets):
        X_m     = df_model[var_set].values.astype(float)
        var_str = " + ".join(
            [c.replace("_lag", " L") for c in var_set]
        )
        try:
            gam_m, X_full_m = fit_gam(time_idx, X_m, y_gam)
            aic, de, pr2    = _safe_gam_stats(gam_m)

            model_results.append({
                "Model_ID" : f"M{idx+1}",
                "Variables": " + ".join(var_set),
                "n_pred"   : len(var_set),
                "AIC"      : round(aic,  3),
                "Dev_Expl" : round(de,   1),
                "Pseudo_R2": round(pr2,  3),
                "_gam"     : gam_m,
                "_cols"    : var_set,
                "_Xfull"   : X_full_m
            })
            print(f"  M{idx+1:<3} {var_str:<58}"
                  f" {aic:>10.2f} {de:>6.1f}% {pr2:>7.3f}")

        except Exception as e:
            print(f"  M{idx+1:<3} {var_str:<58}"
                  f"  FAILED: {str(e)[:35]}")

    # ── Guard ────────────────────────────────────────────────
    if not model_results:
        print(
            "\n[ERROR] All GAM models failed.\n"
            "  Run this diagnostic block in a new cell:\n\n"
            "    from pygam import PoissonGAM, s, l\n"
            "    import numpy as np\n"
            "    t = np.arange(80, dtype=float)\n"
            "    X = np.random.rand(80, 1)\n"
            "    y = np.random.poisson(5, 80).astype(float)\n"
            "    gam = PoissonGAM(s(0)+l(1))\n"
            "    gam.fit(np.column_stack([t,X]), y)\n"
            "    print(gam.statistics_)\n\n"
            "  Then paste the output here so the stats keys\n"
            "  can be mapped correctly.\n"
        )
        raise RuntimeError(
            "All GAM models failed — see diagnostic above."
        )

    # Select best
    res_df   = pd.DataFrame([
        {k: v for k, v in r.items() if not k.startswith("_")}
        for r in model_results
    ])
    best_idx = res_df["AIC"].idxmin()
    best_res = model_results[best_idx]
    best_cols = best_res["_cols"]

    print(f"\n  >> BEST: {best_res['Model_ID']}"
          f"  |  Variables: {best_res['Variables']}"
          f"  |  AIC: {best_res['AIC']}"
          f"  |  Dev: {best_res['Dev_Expl']}%"
          f"  |  R2: {best_res['Pseudo_R2']}")

    res_df.to_csv("multivariate_model_selection.csv", index=False)
    print("\n  Saved: multivariate_model_selection.csv")


# ============================================================
# SECTION 10 — BEST MODEL FULL-DATA PREDICTIONS
# ============================================================

print("\n" + "=" * 65)
print("SECTION 10: BEST MODEL — FULL-DATA PREDICTIONS")
print("=" * 65)

if best_res is None:
    print("  SKIPPED — no GAM model available.")
    df_model["gam_pred"] = np.nan
else:
    best_gam = best_res["_gam"]
    X_best   = best_res["_Xfull"]
    df_model["gam_pred"] = np.clip(
        best_gam.predict(X_best), 0, None
    )
    print(f"\n  Variables : {best_cols}")
    print(f"  AIC       : {best_res['AIC']}")
    print(f"  Dev expl  : {best_res['Dev_Expl']}%")
    print(f"  Pseudo R2 : {best_res['Pseudo_R2']}")


# ============================================================
# SECTION 11 — RESIDUAL DIAGNOSIS ACF / PACF
# ============================================================

print("\n" + "=" * 65)
print("SECTION 11: RESIDUAL DIAGNOSIS — ACF / PACF")
print("=" * 65)

if df_model["gam_pred"].isna().all():
    print("  SKIPPED — no GAM predictions available.")
else:
    resid = (df_model[CASE_COL].values
             - df_model["gam_pred"].fillna(0).values)

    fig, axes = plt.subplots(1, 2, figsize=(12, 4), dpi=150)
    plot_acf(resid, lags=12, ax=axes[0], alpha=0.05)
    axes[0].set_title("ACF — GAM Residuals")
    axes[0].set_xlabel("Lag (months)")
    axes[0].grid(True, alpha=0.3)

    plot_pacf(resid, lags=12, ax=axes[1], alpha=0.05,
              method="ywm")
    axes[1].set_title("PACF — GAM Residuals")
    axes[1].set_xlabel("Lag (months)")
    axes[1].grid(True, alpha=0.3)

    plt.suptitle(
        "Residual Autocorrelation Check — Best Poisson GAM\n"
        "Bars within dashed lines = no autocorrelation",
        fontsize=10, fontweight="bold"
    )
    plt.tight_layout()
    save_fig("residual_acf_pacf.png")
    plt.show()


# ============================================================
# SECTION 12 — 10-FOLD TIME-SERIES CROSS-VALIDATION
# ============================================================

print("\n" + "=" * 65)
print("SECTION 12: 10-FOLD TIME-SERIES CROSS-VALIDATION")
print("=" * 65)

tscv = TimeSeriesSplit(n_splits=N_CV_SPLITS)

cv = {
    "Poisson_GAM"  : {"yt": [], "yp": [], "dt": []},
    "Random_Forest": {"yt": [], "yp": [], "dt": []},
    "SVM"          : {"yt": [], "yp": [], "dt": []}
}

X_ml   = df_model[best_cols].values.astype(float)
t_all  = df_model["time_index"].values.astype(float)
y_all  = df_model[CASE_COL].values.astype(float)
dt_all = df_model["date"].values

print(f"\n  {N_CV_SPLITS}-fold TimeSeriesSplit — no shuffle, "
      f"no data leakage\n")

for fold, (tr, te) in enumerate(tscv.split(X_ml), 1):

    Xtr, Xte = X_ml[tr], X_ml[te]
    ttr, tte = t_all[tr], t_all[te]
    ytr, yte = y_all[tr], y_all[te]
    dte      = dt_all[te]

    # ── Poisson GAM ──────────────────────────────────────────
    if PYGAM_OK:
        try:
            gam_cv, _   = fit_gam(ttr, Xtr, ytr)
            Xte_gam     = np.column_stack(
                [tte.reshape(-1, 1), Xte]
            )
            pred_gam = np.clip(gam_cv.predict(Xte_gam), 0, None)
        except Exception as e:
            print(f"  GAM fold {fold} error ({e}) — mean fallback")
            pred_gam = np.full(len(yte), ytr.mean())
    else:
        pred_gam = np.full(len(yte), ytr.mean())

    cv["Poisson_GAM"]["yt"].extend(yte.tolist())
    cv["Poisson_GAM"]["yp"].extend(pred_gam.tolist())
    cv["Poisson_GAM"]["dt"].extend(dte.tolist())

    # ── Random Forest ────────────────────────────────────────
    rf_cv = RandomForestRegressor(
        n_estimators=500, max_depth=10,
        min_samples_split=5, random_state=RANDOM_STATE
    )
    rf_cv.fit(Xtr, ytr)
    pred_rf = np.clip(rf_cv.predict(Xte), 0, None)

    cv["Random_Forest"]["yt"].extend(yte.tolist())
    cv["Random_Forest"]["yp"].extend(pred_rf.tolist())
    cv["Random_Forest"]["dt"].extend(dte.tolist())

    # ── SVM ──────────────────────────────────────────────────
    sc      = StandardScaler()
    pred_sv = np.clip(
        SVR(kernel="rbf", C=10, gamma="scale", epsilon=0.5
            ).fit(sc.fit_transform(Xtr), ytr
                  ).predict(sc.transform(Xte)),
        0, None
    )

    cv["SVM"]["yt"].extend(yte.tolist())
    cv["SVM"]["yp"].extend(pred_sv.tolist())
    cv["SVM"]["dt"].extend(dte.tolist())

    gam_rmse = np.sqrt(mean_squared_error(yte, pred_gam))
    rf_rmse  = np.sqrt(mean_squared_error(yte, pred_rf))
    sv_rmse  = np.sqrt(mean_squared_error(yte, pred_sv))

    print(f"  Fold {fold:2d} | train={len(tr):4d} | test={len(te):3d}"
          f" | GAM={gam_rmse:7.1f}"
          f" | RF={rf_rmse:7.1f}"
          f" | SVM={sv_rmse:7.1f}")


# ============================================================
# SECTION 13 — MODEL PERFORMANCE METRICS
# ============================================================

print("\n" + "=" * 65)
print("SECTION 13: MODEL PERFORMANCE — OUT-OF-SAMPLE CV")
print("=" * 65)

print(f"\n{'Model':<18} {'RMSE':>8} {'MAE':>8}"
      f" {'R2':>8} {'Pearson_r':>10}")
print("-" * 56)

met_rows = []
for name, store in cv.items():
    m = regression_metrics(store["yt"], store["yp"])
    met_rows.append({"Model": name,
                     **{k: round(v, 3) for k, v in m.items()}})
    print(f"  {name:<16} {m['RMSE']:>8.3f} {m['MAE']:>8.3f}"
          f" {m['R2']:>8.3f} {m['Pearson_r']:>10.3f}")

met_df        = pd.DataFrame(met_rows)
best_cv_model = met_df.loc[met_df["R2"].idxmax(), "Model"]

print(f"\n  >> Best model (R2): {best_cv_model}")
met_df.to_csv("model_performance_metrics.csv", index=False)
print("  Saved: model_performance_metrics.csv")


# ============================================================
# SECTION 14 — OBSERVED VS PREDICTED PLOT
# ============================================================

print("\n" + "=" * 65)
print("SECTION 14: OBSERVED VS PREDICTED PLOT")
print("=" * 65)

rf_full = RandomForestRegressor(
    n_estimators=500, max_depth=10,
    min_samples_split=5, random_state=RANDOM_STATE
)
rf_full.fit(X_ml, y_all)
df_model["rf_full"] = np.clip(rf_full.predict(X_ml), 0, None)

sc_full  = StandardScaler()
svm_full = SVR(kernel="rbf", C=10, gamma="scale", epsilon=0.5)
svm_full.fit(sc_full.fit_transform(X_ml), y_all)
df_model["svm_full"] = np.clip(
    svm_full.predict(sc_full.transform(X_ml)), 0, None
)

fig, ax = plt.subplots(figsize=(14, 6), dpi=150)

ax.plot(df_model["date"], df_model[CASE_COL],
        color="red", linewidth=2, label="Observed", zorder=5)

if not df_model["gam_pred"].isna().all():
    ax.plot(df_model["date"], df_model["gam_pred"],
            color="green", linewidth=1.5,
            label="Poisson GAM (full)", zorder=4)
    cv_dates = pd.to_datetime(cv["Poisson_GAM"]["dt"])
    ax.scatter(cv_dates, cv["Poisson_GAM"]["yp"],
               color="darkgreen", s=18, zorder=6, alpha=0.85,
               label="GAM 10-fold CV")

ax.plot(df_model["date"], df_model["rf_full"],
        color="blue", linewidth=1.2, linestyle="--",
        label="Random Forest (full)", zorder=3)
ax.plot(df_model["date"], df_model["svm_full"],
        color="purple", linewidth=1.2, linestyle=":",
        label="SVM (full)", zorder=2)

ax.set_xlabel("Year", fontsize=11)
ax.set_ylabel("Monthly Dengue Cases", fontsize=11)
ax.set_title(
    "Observed vs Predicted Monthly Dengue Cases\n"
    "Kaski District, Nepal",
    fontsize=12, fontweight="bold"
)
ax.legend(fontsize=9, loc="upper left")
ax.grid(True, alpha=0.3)
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
plt.tight_layout()
save_fig("observed_vs_predicted.png")
plt.show()


# ============================================================
# SECTION 15 — SCATTER PLOTS
# ============================================================

print("\n" + "=" * 65)
print("SECTION 15: SCATTER PLOTS")
print("=" * 65)

pdata = [
    ("Poisson GAM (CV)",
     cv["Poisson_GAM"]["yt"], cv["Poisson_GAM"]["yp"], "green"),
    ("Random Forest (CV)",
     cv["Random_Forest"]["yt"], cv["Random_Forest"]["yp"], "royalblue"),
    ("SVM (CV)",
     cv["SVM"]["yt"], cv["SVM"]["yp"], "mediumpurple"),
]

fig, axes = plt.subplots(1, 3, figsize=(14, 5), dpi=150)

for ax, (name, yt, yp, col) in zip(axes, pdata):
    yta = np.array(yt); ypa = np.array(yp)
    ax.scatter(yta, ypa, alpha=0.6, color=col,
               edgecolors="none", s=25)
    lim = [min(yta.min(), ypa.min()) - 1,
           max(yta.max(), ypa.max()) + 1]
    ax.plot(lim, lim, "k--", linewidth=0.9, label="1:1 line")
    m = regression_metrics(yta, ypa)
    ax.set_xlabel("Observed Cases")
    ax.set_ylabel("Predicted Cases" if "GAM" in name else "")
    ax.set_title(
        f"{name}\nr={m['Pearson_r']:.3f}  R²={m['R2']:.3f}"
    )
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)

fig.suptitle(
    "Observed vs Predicted (10-fold CV) — Kaski District",
    fontsize=11, fontweight="bold"
)
plt.tight_layout()
save_fig("scatter_obs_pred.png")
plt.show()


# ============================================================
# SECTION 16 — RANDOM FOREST FEATURE IMPORTANCE
# ============================================================

print("\n" + "=" * 65)
print("SECTION 16: RANDOM FOREST FEATURE IMPORTANCE")
print("=" * 65)

imp = pd.Series(
    rf_full.feature_importances_, index=best_cols
).sort_values(ascending=True)

fig, ax = plt.subplots(
    figsize=(7, max(3, len(best_cols) * 0.7)), dpi=150
)
imp.plot(kind="barh", ax=ax, color="steelblue",
         edgecolor="black", linewidth=0.5)
ax.set_xlabel("Importance Score")
ax.set_title("Random Forest Feature Importance\nKaski District")
ax.grid(True, alpha=0.3, axis="x")
plt.tight_layout()
save_fig("feature_importance.png")
plt.show()

print("\n  Ranking:")
for feat, val in imp.sort_values(ascending=False).items():
    print(f"    {feat:<30}: {val:.4f}")


# ============================================================
# SECTION 17 — MODEL COMPARISON CHART
# ============================================================

print("\n" + "=" * 65)
print("SECTION 17: MODEL COMPARISON CHART")
print("=" * 65)

models_  = met_df["Model"].tolist()
metrics_ = ["RMSE", "MAE", "R2", "Pearson_r"]
x        = np.arange(len(models_))
bar_cols = ["steelblue", "tomato", "mediumseagreen", "mediumpurple"]

fig, axes = plt.subplots(1, 4, figsize=(16, 4), dpi=150)

for ax, metric, col in zip(axes, metrics_, bar_cols):
    vals = met_df[metric].values
    bars = ax.bar(x, vals, color=col, edgecolor="black",
                  linewidth=0.5, width=0.5)
    for bar, val in zip(bars, vals):
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.005 * max(abs(vals)),
            f"{val:.3f}", ha="center", va="bottom", fontsize=8
        )
    ax.set_xticks(x)
    ax.set_xticklabels(
        [m.replace("_", "\n") for m in models_], fontsize=8
    )
    ax.set_title(metric, fontweight="bold")
    ax.grid(True, alpha=0.3, axis="y")
    if metric in ["R2", "Pearson_r"]:
        ax.set_ylim(0, 1.15)

fig.suptitle(
    "Model Performance Comparison (10-fold CV) — Kaski District",
    fontsize=11, fontweight="bold"
)
plt.tight_layout()
save_fig("model_comparison_chart.png")
plt.show()


# ============================================================
# SECTION 18 — RISK CLASSIFICATION
# ============================================================

print("\n" + "=" * 65)
print("SECTION 18: DENGUE RISK CLASSIFICATION")
print("=" * 65)

if df_model["gam_pred"].isna().all():
    # Fall back to RF if GAM unavailable
    pred_col = "rf_full"
    print("  Using RF predictions for risk classification "
          "(GAM unavailable).")
else:
    pred_col = "gam_pred"

q33 = np.percentile(df_model[pred_col].values, 33)
q66 = np.percentile(df_model[pred_col].values, 66)

df_model["risk_class"] = np.where(
    df_model[pred_col] <= q33, 1,
    np.where(df_model[pred_col] <= q66, 2, 3)
)
df_model["risk_label"] = df_model["risk_class"].map(
    {1: "Low", 2: "Medium", 3: "High"}
)

print(f"\n  Low    : <= {q33:.2f} cases/month")
print(f"  Medium : {q33:.2f} – {q66:.2f}")
print(f"  High   : > {q66:.2f}")
print(f"\n  Distribution:")
print(df_model["risk_label"].value_counts().to_string())


# ============================================================
# SECTION 19 — SAVE ALL RESULTS
# ============================================================

print("\n" + "=" * 65)
print("SECTION 19: SAVING RESULTS")
print("=" * 65)

keep = (
    ["date", YEAR_COL, MONTH_COL, "time_index", CASE_COL]
    + best_cols
    + ["month_sin", "month_cos",
       "gam_pred", "rf_full", "svm_full",
       "risk_class", "risk_label"]
)
keep = [c for c in keep if c in df_model.columns]
df_model[keep].to_csv("dengue_kaski_final_results.csv",
                       index=False)

outputs = [
    "dengue_kaski_final_results.csv",
    "model_performance_metrics.csv",
    "multivariate_model_selection.csv",
    "univariate_model_results.csv",
    "crosscorrelation_results.csv",
    "predictor_correlation_matrix.csv",
    "descriptive_statistics.csv",
    "timeseries_plot.png",
    "crosscorrelation_plot.png",
    "correlation_heatmap_predictors.png",
    "residual_acf_pacf.png",
    "observed_vs_predicted.png",
    "scatter_obs_pred.png",
    "feature_importance.png",
    "model_comparison_chart.png",
]
print("\n  Files:")
for f in outputs:
    print(f"    {f}")

print("\n" + "=" * 65)
print("ANALYSIS COMPLETE")
print(f"  Primary model : Poisson GAM (Acharya et al. 2018)")
print(f"  Comparative   : Random Forest  +  SVM")
print(f"  Validation    : {N_CV_SPLITS}-fold TimeSeriesSplit")
print(f"  Best CV model : {best_cv_model}")
print("=" * 65)
```
