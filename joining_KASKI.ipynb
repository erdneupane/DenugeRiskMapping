```python
import pandas as pd
```

```python
# Load dengue cases table
dengue_df = pd.read_csv(r"C:\Users\dhrub\Desktop\Data\kaski_monthly_dengue_2019_2025.csv")

# Load environmental variables table
env_df = pd.read_csv(r"C:\Users\dhrub\Desktop\Data\kaski_env_monthly_2019.csv")

# Merge using year and month (keep all env rows)
merged_df = pd.merge(
    env_df,
    dengue_df,
    on=["year", "month"],
    how="left"
)

# Fill missing dengue cases with 0
merged_df["total_cases"] = merged_df["total_cases"].fillna(0).astype(int)

# Save output
merged_df.to_csv(r"C:\Users\dhrub\Desktop\Data\kaski_env_dengue_merged.csv", index=False)

print("Merged file saved successfully!")
print(merged_df.head())
```
